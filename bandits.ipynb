# COIN EXAMPLE

```python
import random

cups = {
    "A": {
        "coins": [0, 1],
        "probs": [0.4, 0.6]
    },
    "B": {
        "coins": [0, 1],
        "probs": [0.3, 0.7]
    },
    "C": {
        "coins": [0, 1],
        "probs": [0.5, 0.5]
    }
}

# Statistics
totals = {"A": 0, "B": 0, "C": 0}
counts = {"A": 0, "B": 0, "C": 0}


def reward(cup):
    return random.choices(
        cups[cup]["coins"],
        weights=cups[cup]["probs"],
        k=1
    )[0]


while True:
    choice = input("Choose cup (A/B/C) or q: ").upper()

    if choice == "Q":
        break

    if choice not in cups:
        print("Invalid cup")
        continue

    coin = reward(choice)

    totals[choice] += coin
    counts[choice] += 1

    averages = {}
    for cup in ["A", "B", "C"]:
        if counts[cup] > 0:
            averages[cup] = totals[cup] / counts[cup]
        else:
            averages[cup] = 0

    #print("\nCup Results")
    print(" A    B    C")

    values = {
        "A": "",
        "B": "",
        "C": ""
    }
    values[choice] = str(coin)

    print(f"{values['A']:^4} {values['B']:^4} {values['C']:^4}")

    print("\nAverage Reward")
    print(
        f"{averages['A']:^4.2f} "
        f"{averages['B']:^4.2f} "
        f"{averages['C']:^4.2f}"
    )
    print()
```

# Dice Example

```python
import random

# Three dice
dice = {
    "A": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [0.30, 0.25, 0.20, 0.15, 0.07, 0.03]
    },
    "B": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [0.05, 0.10, 0.15, 0.20, 0.25, 0.25]
    },
    "C": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [1/6] * 6
    }
}

# Statistics
totals = {"A": 0, "B": 0, "C": 0}
counts = {"A": 0, "B": 0, "C": 0}

history_rows = []

def roll_die(die):
    return random.choices(
        dice[die]["sides"],
        weights=dice[die]["probs"],
        k=1
    )[0]

while True:
    choice = input("Choose die (A/B/C) or q: ").upper()

    if choice == "Q":
        break

    if choice not in dice:
        print("Invalid die")
        continue

    roll = roll_die(choice)

    totals[choice] += roll
    counts[choice] += 1

    averages = {}
    for die in ["A", "B", "C"]:
        if counts[die] > 0:
            averages[die] = totals[die] / counts[die]
        else:
            averages[die] = 0

    # after rolling
    if choice == "A":
        history_rows.append((roll, "-", "-"))
    elif choice == "B":
        history_rows.append(("-", roll, "-"))
    else:
        history_rows.append(("-", "-", roll))

    print("\n   A        B        C")
    for a, b, c in history_rows:
        print(f"{str(a):^8} {str(b):^8} {str(c):^8}")
    print(
        f"A={averages['A']:.2f}  "
        f"B={averages['B']:.2f}  "
        f"C={averages['C']:.2f}"
    )

    print("-" * 40)
```

# Dice Example: Uniform Random Policy

```python
import random
import time

# Three dice
dice = {
    "A": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [0.30, 0.25, 0.20, 0.15, 0.07, 0.03]
    },
    "B": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [0.05, 0.10, 0.15, 0.20, 0.25, 0.25]
    },
    "C": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [1/6] * 6
    }
}

# Statistics
totals = {"A": 0, "B": 0, "C": 0}
counts = {"A": 0, "B": 0, "C": 0}

history_rows = []

def roll_die(die):
    return random.choices(
        dice[die]["sides"],
        weights=dice[die]["probs"],
        k=1
    )[0]

# Random policy: 100 rolls
for step in range(1, 101):

    # Choose a die uniformly at random
    choice = random.choice(["A", "B", "C"])

    roll = roll_die(choice)

    totals[choice] += roll
    counts[choice] += 1

    # after rolling
    if choice == "A":
        history_rows.append((roll, "-", "-"))
    elif choice == "B":
        history_rows.append(("-", roll, "-"))
    else:
        history_rows.append(("-", "-", roll))

    print("\n   A        B        C")
    print(f"{history_rows[-1][0]:^8} {history_rows[-1][1]:^8} {history_rows[-1][2]:^8}")

    time.sleep(0.5)   # pause for half a second

averages = {}
for die in ["A", "B", "C"]:
    if counts[die] > 0:
        averages[die] = totals[die] / counts[die]
    else:
        averages[die] = 0
print(
    f"A={averages['A']:.2f}  "
    f"B={averages['B']:.2f}  "
    f"C={averages['C']:.2f}"
    )

print("-" * 40)
```

# Dice Example: Greedy Policy

```python
import random
import time

# Three dice
dice = {
    "A": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [0.30, 0.25, 0.20, 0.15, 0.07, 0.03]
    },
    "B": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [0.05, 0.10, 0.15, 0.20, 0.25, 0.25]
    },
    "C": {
        "sides": [1, 2, 3, 4, 5, 6],
        "probs": [1/6] * 6
    }
}

# Statistics
totals = {"A": 0, "B": 0, "C": 0}
counts = {"A": 0, "B": 0, "C": 0}

history_rows = []

def roll_die(die):
    return random.choices(
        dice[die]["sides"],
        weights=dice[die]["probs"],
        k=1
    )[0]

averages = {}
for die in ["A", "B", "C"]:
    counts[die] = 0

# Random policy: 100 rolls
for step in range(1, 101):

    if step == 1:
        # Choose a die uniformly at random
        choice = random.choice(["A", "B", "C"])
    else:
        # Choose a die uniformly at random
        choice = max(averages, key=averages.get)

    roll = roll_die(choice)

    totals[choice] += roll
    counts[choice] += 1

    # after rolling
    if choice == "A":
        history_rows.append((roll, "-", "-"))
    elif choice == "B":
        history_rows.append(("-", roll, "-"))
    else:
        history_rows.append(("-", "-", roll))

    print("\n   A        B        C")
    print(f"{history_rows[-1][0]:^8} {history_rows[-1][1]:^8} {history_rows[-1][2]:^8}")

    time.sleep(0.5)   # pause for half a second

    for die in ["A", "B", "C"]:
        if counts[die] > 0:
            averages[die] = totals[die] / counts[die]
        else:
            averages[die] = 0
print(
    f"A={averages['A']:.2f}  "
    f"B={averages['B']:.2f}  "
    f"C={averages['C']:.2f}"
    )

print("-" * 40)
```
